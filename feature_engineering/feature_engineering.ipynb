## Feature engineering notebook

Derive the full dataset—with hand-engineered and LDA-derived title/description features—from used laptop listings data pulled from the eBay API. 

Start with the abridged dataset (= no title/description features). First we get the hand-engineered features.

```python
import call_abridged_data

laptopData = call_abridged_data.read_data()
laptopData.head()
```

```python
import get_hand_engineered_features

laptopData = get_hand_engineered_features.get_year_mention_features(laptopData)
laptopData.head()
```

```python
laptopData = get_hand_engineered_features.get_size_mention_features(laptopData)
laptopData = get_hand_engineered_features.get_memory_mention_features(laptopData)
laptopData.head()
```

```python
laptopData = get_hand_engineered_features.get_adjective_mention_features(laptopData)
laptopData = get_hand_engineered_features.get_length_features(laptopData)
laptopData.head()
```

Apply LDA to laptop descriptions, get some topics, have columns quantifying how much a description belongs to each topic.

```python
import get_LDA_topic_features

laptopTopics = get_LDA_topic_features.LDA(laptopData)
```

```python
laptopTopics.head()
```
